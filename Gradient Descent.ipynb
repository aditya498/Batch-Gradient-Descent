```python
import numpy as np
from sklearn import preprocessing
```

```python
def step_gradient(x,y, learning_rate, m):
    M = len(x)
    new_m=np.array([])
#     print(x.shape,y.shape,m.shape)
    for j in range(len(m)):
        temp=0
        for i in range(M):
            temp+=(y[i][0]-(x[i,:]*m).sum())*x[i,j]*2
        temp=temp/M
        temp=-temp
        new_m=np.append(new_m,temp)
        
    for i in range(len(new_m)):
        new_m[i]=m[i]-learning_rate*new_m[i]
    return new_m     


def cost(x,y, m):
    M = len(x)
    cost=0
    for i in range(M):
        cost+=((y[i][0]-(x[i,:]*m).sum())**2)
    cost=cost/M  
    return cost


def gd(x,y, learning_rate, num_iterations):
    m = np.zeros(x.shape[1])
    for i in range(num_iterations):
        m= step_gradient(x,y, learning_rate, m)
        c=cost(x,y, m)
        print(i, " Cost: ", c)
    return m


scaler=preprocessing.StandardScaler()


def run():
    points = np.loadtxt("/Users/adityamittal/Downloads/0000000000002417_training_boston_x_y_train.csv", delimiter=",")
    x=points[:,:13]
    y=points[:,13]
    y=y.reshape(-1,1)
    temp=np.ones(points.shape[0]).reshape(-1,1)
    scaler.fit(x)
    x=scaler.transform(x)
    x=np.append(x,temp,axis=1)
    learning_rate = 0.15
    num_iterations = 1000
    m= gd(x,y, learning_rate, num_iterations)
    return m
    
```

```python
m=run()
```

```python
def predict(test,m):
    output=np.array([])
    for i in range(len(test)):
        temp=(test[i,:]*m[:-1]).sum()+m[-1]
        output=np.append(output,temp)
    return output
```

```python
test = np.loadtxt("/Users/adityamittal/Downloads/0000000000002417_test_boston_x_test.csv", delimiter=",")
test=scaler.transform(test)
```

```python
pred=predict(test,m)
```
